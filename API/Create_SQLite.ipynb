```python
#!/usr/bin/env python3

import sqlite3
import pandas as pd

from pathlib import Path

city_path = "C:/Users/smoot/anaconda3/Bootcamp/Project3/data/city.sqlite"
Path(city_path).touch()

conn = sqlite3.connect(city_path)
c = conn.cursor()

c.execute('''CREATE TABLE IF NOT EXISTS "city" (
    "City" TEXT NOT NULL,
    "State" TEXT NOT NULL,
    "Population" INTEGER,
    "CBSA Code" INTEGER,
    "CO" NUMERIC,
    "Pb" NUMERIC,
    "NO2 (AM)" NUMERIC,
    "NO2 (1hr)" NUMERIC,
    "O3" NUMERIC,
    "PM10" NUMERIC,
    "PM2.5 (WAM)" NUMERIC,
    "PM2.5 (24hr)" NUMERIC,
    "SO2" NUMERIC,
    PRIMARY KEY ("City", "State")
)''')

csv_city = pd.read_csv("C:/Users/smoot/anaconda3/Bootcamp/Project3/data/AirQuality_City_2022.csv")
csv_city.to_sql("city", conn, if_exists='append', index=False)

conn.close()
```

```python
county_path = "C:/Users/smoot/anaconda3/Bootcamp/Project3/data/county.sqlite"
Path(county_path).touch()

conn = sqlite3.connect(county_path)
c = conn.cursor()

c.execute('''CREATE TABLE IF NOT EXISTS "county" (
    "State" TEXT NOT NULL,
    "County" TEXT NOT NULL,
    "County Code" INTEGER,
    "Population" INTEGER,
    "CO" NUMERIC,
    "Pb" NUMERIC,
    "NO2 (AM)" NUMERIC,
    "NO2 (1hr)" NUMERIC,
    "O3" NUMERIC,
    "PM10" NUMERIC,
    "PM2.5 (WAM)" NUMERIC,
    "PM2.5 (24hr)" NUMERIC,
    "SO2" NUMERIC,
    PRIMARY KEY ("State", "County")
)
''')

csv_county = pd.read_csv("C:/Users/smoot/anaconda3/Bootcamp/Project3/data/ctyfactbook2022.csv")
csv_county.to_sql("county", conn, if_exists='append', index=False)

conn.close()
```
